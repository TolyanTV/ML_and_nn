```python
# разбейте исходную таблицу на 2: по полу пассажиров
# удалите пустые строки или заполните их соответствующими символами
# Построить гистограмму распределения пассажиров по возрасту
# Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола
# визуализировать возраст при помощи различных настроек метода plot
```

```python
import pandas as pd

data = pd.read_csv("titanic.csv")
data
```

```python
# разбейте исходную таблицу на 2: по полу пассажиров
data_male = data[data['Sex'] == 'male']
data_male
```

```python
# разбейте исходную таблицу на 2: по полу пассажиров
data_female = data[data['Sex'] == 'female']
data_female
```

```python
#удалите пустые строки или заполните их соответствующими символами
data = data.dropna()
data_male = data_male.dropna()
data_female = data_female.dropna()
data
```

```python
# Построить гистограмму распределения пассажиров по возрасту

import matplotlib.pyplot as plt

plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.show()
```

```python
# Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола
def func(data):
    return {
        'min': data['Age'].min(),
        'max': data['Age'].max(),
        'mean': data['Age'].mean(),
        'mode': data['Age'].mode().iloc[0],
        'median': data['Age'].median()
    }

pd.DataFrame({
    'Male': func(data_male),
    'Female': func(data_female)
})
```

```python
# визуализировать возраст при помощи различных настроек метода plot
plt.figure(figsize=(10, 6))

data_male['Age'].plot(kind='density', label='Возрастная линия плотность для мужчин', color='blue')
data_female['Age'].plot(kind='density', label='Возрастная линия плотность для женщин', color='red')
plt.xlabel('Возраст')
plt.title('График плотности по возрасту и полу')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
data.boxplot(column='Age', by='Sex')
plt.title('Ящик с усами по возрасту и полу')
plt.suptitle('')
plt.xlabel('Пол')
plt.ylabel('Возраст')
plt.show()


plt.scatter(data_male.index, data_male['Age'], color='blue', alpha=0.5)
plt.scatter(data_female.index, data_female['Age'], color='red', alpha=0.5)

plt.title('Точечная диаграмма')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
